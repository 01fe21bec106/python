sklearn and skit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([2,13,24,38,41,57]).reshape((-1,1))
y=np.array([3,23,5,67,78,22,])
plt.xlabel("IV")
plt.ylabel("DV")
plt.plot(x,y)
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color="green",marker='*')
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
print("MSE",mse)
print("intercept",SLR.intercept_)
print("Slope",SLR.coef_)
```

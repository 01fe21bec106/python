```python

def read_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = int(input("Enter element at position ({i+1}, {j+1}): "))
            row.append(element)
        matrix.append(row)
    return matrix
def print_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))
def multiply_matrices(a, b):
    rows1 = len(a)
    cols1 = len(a[0])
    rows2 = len(b)
    cols2 = len(b[0])

    if cols1 != rows2:
        return "Matrix multiplication not possible. Number of columns in a must be equal to the number of rows in b."

    result = [[0 for _ in range(cols2)] for _ in range(rows1)]

    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += a[i][k] * b[k][j]

    return result
print("Enter the first matrix (3x2):")
a = read_matrix(3, 2)

print("Enter the second matrix (2x3):")
b = read_matrix(2, 3)
result_matrix = multiply_matrices(a, b)
if isinstance(result_matrix, str):
    print(result_matrix)
else:
    print("Matrix Multiplication Result:")
    print_matrix(result_matrix)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([1,3,5,6,19]).reshape((-1,1))
y=np.array([43,55,77,68,12])
plt.plot(x,y)
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='green')
plt.show()

print("MSE",mse)
print("slope:",SLR.coef_)
print("intercept:",SLR.intercept_)

```
